```python
import sys
from collections import deque

N = 4

# ↑, ↖, ←, ↙, ↓, ↘, →, ↗
s_dir = [[-1,0],[-1,-2],[0,-2],[1,-2],[1,0],[1,2],[0,2],[-1,2]]

fish_arr = []
for _ in range(N):
    # for jupyter
    fish_arr.append(list(map(int, input().split())))
    
#     for bj
#     fish_arr.append(list(map(int, sys.stdin.readline().split())))
    
def fish_moving(area):
    queue = []
    visited = [0] * 18
    
    for fr_idx in range(N):
        for fc_idx in range(0,N*2,2):
            queue.append([area[fr_idx][fc_idx], fr_idx, fc_idx])
    
    queue = deque(sorted(queue, key=lambda arr: arr[0]))
    while queue:
        # 이미 2배 곱해진 것
        fish_num, r_idx, c_idx = queue.popleft()
        
        if fish_num == -1:
            continue
        
        if area[r_idx][c_idx] == 18:
            break
            
        if visited[fish_num - 1] == 1:
            queue.append([fish_num, r_idx, c_idx])
            continue
        
        else:
            visited[fish_num - 1] = 1
            dir_num = area[r_idx][c_idx + 1] - 1

            nr = r_idx + s_dir[dir_num][0]
            nc = c_idx + s_dir[dir_num][1]
            nd = dir_num + 1
            
            rot_cnt = 0
            # 맞는 방향을 못찾았을 때 대비 처리
            for dir_idx in range(1, len(s_dir)+1):
                if nr >= N or nr < 0 or nc >= 2*N or nc < 0 or area[nr][nc] == 18:
                    # 위치 45도 돌리기
                    nr = r_idx + s_dir[(dir_num + dir_idx) % 8][0]
                    nc = c_idx + s_dir[(dir_num + dir_idx) % 8][1]
                    nd = ((dir_num + dir_idx) % 8) + 1
                    rot_cnt += 1
                else:
                    # 탈출
                    break
                    
            if rot_cnt == len(s_dir):
                continue
            
            # 기존 위치가 변경되었을 시 새로운것 append, 기존의 데이터는 -1로 지나침   
            queue[queue.index([area[nr][nc], nr, nc])][0] = -1
            
            queue.append([fish_num, nr, nc])
            queue.append([area[nr][nc], r_idx, c_idx])
        
            area[nr][nc], area[r_idx][c_idx] = fish_num, area[nr][nc]
            area[nr][nc+1], area[r_idx][c_idx+1] = nd , area[nr][nc+1]
        
            queue = deque(sorted(queue, key=lambda arr: arr[0]))         
            
    return area

# 상어 queue init
s_queue = deque()
s_queue.append([0, 0, 0, 0, fish_arr])

max_eaten = 0
while s_queue:
    sr_idx, sc_idx, sd_idx, eaten_sum, sea = s_queue.popleft()
    if sea[sr_idx][sc_idx] == 0:
        continue
    
    if max_eaten < eaten_sum:
        max_eaten = eaten_sum
    
    eaten_sum += sea[sr_idx][sc_idx]
    sea[sr_idx][sc_idx] = 18
    
    sd_idx = sea[sr_idx][sc_idx + 1]
    sea = fish_moving(sea)
    
    for loc_idx in range(N):
        nr_s = sr_idx + (s_dir[sd_idx-1][0]) * (loc_idx + 1)
        nc_s = sc_idx + (s_dir[sd_idx-1][1]) * (loc_idx + 1)
        print(nr_s, nc_s)
        
        if nr_s >= N or nr_s < 0 or nc_s >= N or nc_s < 0:
            break

        else:
            sea[sr_idx][sc_idx] = 0
            print(s_queue)
            s_queue.append([nr_s, nc_s, sd_idx, eaten_sum, sea])

print(max_eaten)
```
