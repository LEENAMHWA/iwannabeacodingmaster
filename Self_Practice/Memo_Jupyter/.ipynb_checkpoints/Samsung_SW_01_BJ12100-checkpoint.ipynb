```python
from collections import deque
import sys

def rotate_90(arr, rows):
    new_arr = []
    for r_idx in range(rows):
        row_arr = []
        for c_idx in range(rows):
            row_arr.append(arr[(rows-1)-c_idx][r_idx])
        new_arr.append(row_arr)
    
    return new_arr
# jupyter
N = int(input())

# baekjoon
# N = int(sys.stdin.readline().split()[0])

game = []
for r_idx in range(N):
    # jupyter
    game.append(list(map(int, input().split())))
    
#     # baekjoon
#     game.append(list(map(int, sys.stdin.readline().split())))
# print(game)
    
# rotate
cnt = 0

scn = deque()
scn.append([game,cnt])

iter = 0
while iter < 4:
    tmp_game, n_try = scn.popleft()
    print(tmp_game)
    tmp_game = rotate_90(tmp_game, N)
    
    scn.append([tmp_game, N])
    print(scn)
    iter += 1
        
print(len(scn))        
```
